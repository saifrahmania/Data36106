## 3. Find Optimal Number of Clusters

Note: If you are stuck in previous steps, you can dowload the content of X here: 'https://raw.githubusercontent.com/aso-uts/labs_datasets/main/36106-mlaa/lab07/ex1/df_cleaned.csv'

**[3.1]** Create an empty dataframe and list called respectively 'cluster_df' and 'inertia_list'

```python
# Placeholder for student's code
```

```python
# Solution
cluster_df = pd.DataFrame()
inertia_list = []
```

**[3.2]** Create a new column called 'nb_cluster' on 'cluster_df' that will contain values from 1 to 20

```python
# Placeholder for student's code
```

```python
# Solution
cluster_df['nb_cluster'] = range(2, 21)
cluster_df
```

**[3.3]** Import KMeans from sklearn.cluster

```python
# Placeholder for student's code
```

```python
# Solution
from sklearn.cluster import KMeans
```

**[3.4]** Iterate through the range of cluster and instantiate a kmeans accordingly and store the inertia score

```python
for k in cluster_df['nb_cluster']:
    kmeans = KMeans(n_clusters=k, random_state=8).fit(X)
    inertia_list.append(kmeans.inertia_)
```

**[3.5]** Create a new column called 'inertia' on 'cluster_df' that will contain the inertia values

```python
# Placeholder for student's code
```

```python
# Solution
cluster_df['inertia'] = inertia_list
```

```python
cluster_df
```

**[3.6]** Import altair package

```python
# Placeholder for student's code
```

```python
# Solution
import altair as alt
```

**[3.7]** Plot a line chart that will show the inertia values from the number of clusters

```python
# Placeholder for student's code
```

```python
# Solution
alt.Chart(cluster_df).mark_line().encode(alt.X('nb_cluster'), alt.Y('inertia'))
```

**[3.8]** Create a variable called k that will take the value 5

```python
# Placeholder for student's code
```

```python
# Solution
k = 5
```

## 4. Train a K-means Model

**[4.1]** Instantiate a kmeans with random_state=1, k as n_clusters

```python
# Placeholder for student's code
```

```python
# Solution
kmeans = KMeans(random_state=1, n_clusters=k)
```

**[4.2]** Fit the model with the data

```python
# Placeholder for student's code
```

```python
# Solution
kmeans.fit(X)
```

**[4.3]** Predict the cluster on the data and save the results back to the original dataframe df in a new column called 'kmeans'

```python
# Placeholder for student's code
```

```python
# Solution
df['kmeans'] = kmeans.predict(X)
```

```python
df
```

**[4.4]** Create a list of all the pairs from the features

```python
pairs = [(a, b) for idx, a in enumerate(feat_list) for b in feat_list[idx + 1:]]
pairs
```

**[4.5]** Print scatter plots to visualise the clusters



```python
base = alt.Chart(df).mark_point().encode(color='kmeans:N')
for a, b in pairs:
  base.encode(x=a, y=b).display()
```

**[4.6]** Display the average of all features for each group

```python
df[feat_list + ['kmeans']].groupby('kmeans').mean()
```
