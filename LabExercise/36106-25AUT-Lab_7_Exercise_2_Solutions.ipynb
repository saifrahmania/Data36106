## 3. Train a mean-shift Model

**[3.1]** Import MeanShift from sklearn.cluster

```python
# Placeholder for student's code
```

```python
# Solution
from sklearn.cluster import MeanShift
```

**[3.2]** Instantiate a MeanShift class

```python
# Placeholder for student's code
```

```python
# Solution
meanshift = MeanShift()
```

**[3.3]** Fit the model with the data

```python
# Placeholder for student's code
```

```python
# Solution
meanshift.fit(X)
```

**[3.3]** Predict the cluster on the data and save the results back to the original dataframe df in a new column called `meanshift`

```python
# Placeholder for student's code
```

```python
# Solution
df['meanshift'] = meanshift.predict(X)
```

**[3.4]** Create a list of all the pairs from the features

```python
# Placeholder for student's code
```

```python
# Solution
feat_list = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
pairs = [(a, b) for idx, a in enumerate(feat_list) for b in feat_list[idx + 1:]]
```

**[3.5]** Print scatter plots to visualise the clusters

```python
# Placeholder for student's code
```

```python
# Solution
base = alt.Chart(df).mark_point().encode(color='meanshift:N')
for a, b in pairs:
  base.encode(x=a, y=b).display()
```

**[3.6]** Display the average of all features for each group

```python
# Placeholder for student's code
```

```python
# Solution
df[feat_list + ['meanshift']].groupby('meanshift').mean()
```
