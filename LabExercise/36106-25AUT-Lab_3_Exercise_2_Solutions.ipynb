### 2. Train Default KNN Model

**[2.1]** Import the RMSE and MAE from sklearn

```python
# Placeholder for student's code
```

```python
# Solution
from sklearn.metrics import root_mean_squared_error as rmse
from sklearn.metrics import mean_absolute_error as mae
```

**[2.2]** Import the KNeighborsRegressor module from sklearn

```python
# Placeholder for student's code
```

```python
# Solution
from sklearn.neighbors import KNeighborsRegressor
```

**[2.3]** Instantiate a KNN model into a variable called `knn_default_reg`

```python
# Placeholder for student's code
```

```python
# Solution
knn_default_reg = KNeighborsRegressor()
```

**[2.4]** Fit `knn_default_reg` on the training set

```python
# Placeholder for student's code
```

```python
# Solution
knn_default_reg.fit(X_train, y_train)
```

**[2.5]** Use the trained model to predict the outcome on `X_train` and save them into `y_preds`

```python
# Placeholder for student's code
```

```python
# Solution
y_train_preds = knn_default_reg.predict(X_train)
```

**[2.6]** Display the RMSE and MAE scores on the training set

```python
# Placeholder for student's code
```

```python
# Solution
print(rmse(y_train, y_train_preds))
print(mae(y_train, y_train_preds))
```

**[2.7]** Save the predictions on the validation set into a variable called `y_val_preds` and display the RMSE and MAE scores

```python
# Placeholder for student's code
```

```python
# Solution
y_val_preds = knn_default_reg.predict(X_val)
print(rmse(y_val, y_val_preds))
print(mae(y_val, y_val_preds))
```

**[2.8]** Display the predictions versus the target line charts on the training set

```python
# Placeholder for student's code
```

```python
# Solution
perfect_test = alt.Chart(y_train).mark_line(color='green').encode(
    x='selling_price',
    y='selling_price'
)

df_train = y_train.copy()
df_train['preds'] = y_train_preds

pred_chart_test = alt.Chart(df_train).mark_line().encode(
    x='selling_price',
    y='preds'
  )

pred_chart_test + perfect_test
```

**[2.9]** Display the predictions versus the target line charts on the validation set

```python
# Placeholder for student's code
```

```python
# Solution
perfect_val = alt.Chart(y_val).mark_line(color='green').encode(
    x='selling_price',
    y='selling_price'
)

df_val = y_val.copy()
df_val['preds'] = y_val_preds

pred_chart_val = alt.Chart(df_val).mark_line().encode(
    x='selling_price',
    y='preds'
  )

pred_chart_val + perfect_val
```
